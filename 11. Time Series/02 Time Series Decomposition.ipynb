# 1. Import Library

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from statsmodels.tsa.seasonal import seasonal_decompose,STL
```

# 2. Time Series Decomposition
Teknik untuk memisahkan time series menjadi trend, seasonal, dan residual menggunakan Moving Average.

Ada 2 cara dekomposisinya, diantaranya :
1. additive (`trend+seasonal+residual`) : dipakai pada trend dan seasonal yang tidak terlalu bervariasi, <br>penambahan/pengurangannya cenderung konstan
2. multiplicative (`trend*seasonal*residual`) : dipakai pada trend dan seasonal yang berubah sering berjalannya waktu

Contoh :
1. additive : terdapat peningkatan pelanggan 10000 orang per tahun
2. multiplicative : terdapat peningkatan pelanggan 10% per tahun

# 3. Decompose by Seasonal Decompose

## 3.1. Additive Example

```python
path_loc = '/content/drive/MyDrive/Course Files/Course-Net/02. Kampus Merdeka/11. Time Series/'
df = pd.read_csv(path_loc+'cheese_fred.csv',index_col='date',parse_dates=['date'])

series = df.total_cheese_production
series.plot(figsize=(12,3))
```

Karena trend data bergerak secara konstan, dan ukuran seasonalnya tampak statis, maka kita pilih model = `'additive'`.

```python
result = seasonal_decompose(series,model='additive')
result.plot();
```

## 3.2. Multiplicative Example

```python
df = pd.read_csv(path_loc+'airline.csv',index_col = 'Date', parse_dates=['Date'])
series = df.passengers

series.plot(figsize=(12,3))
```

Pola/ukuran seasonalnya semakin lama semakin membesar sehingga modelnya cocok dengan yang `'multiplicative'`.

```python
result = seasonal_decompose(series,model='mul')
result.plot();
```

# 4. Decompose by STL
Kalau pakai STL, kita nggak perlu state apakah pakai "additive" atau "multiplicative".

```python
df = pd.read_csv(path_loc+'airline.csv',index_col = 'Date', parse_dates=['Date'])
series = df.passengers

result = STL(series).fit()
result.plot()
```
