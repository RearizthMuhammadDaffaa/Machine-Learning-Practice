Buatkan function untuk menghitung luas lingkaran dengan formula di bawah ini.

$L=\pi\times r^2$

```python
def luas_circle(r):
   L = 3.14*r**2
   return L
```

```python
# function dengan argumen yang punya default parameter
def luas_circle(r,phi=22/7):
   L = phi*r**2
   return L

print(luas_circle(7,3.14))
print(luas_circle(phi=3.14,r=7))
```

```python
# function untuk menghitung volume balok dengan argumen default berupa l dan t
def volume_luas_balok(p, l=4, t=5):
    V = p * l * t
    LP = 2*(p*l + p*t + l*t)
    return V,LP
volume_luas_balok(7)
```

```python
volume_luas_balok(7)[0]
```
