```python
from nltk.tokenize import word_tokenize,sent_tokenize
from nltk.corpus import stopwords
from string import punctuation

# !pip install Sastrawi
from Sastrawi.StopWordRemover.StopWordRemoverFactory import StopWordRemoverFactory
from Sastrawi.Stemmer.StemmerFactory import StemmerFactory
```

# 2. Tokenisasi
proses breakdown kalimat/paragraf menjadi bagian-bagiannya yang lebih kecil seperti :
1. kalimat menjadi kata demi kata
2. paragraf menjadi kalimat demi kalimat


```python
text = "Hai, kami sedang belajar DS di Course-Net. Mau join?"
```

```python
## word_tokenize
word_tokenize(text.lower())
```

```python
## sentence tokenize
sent_tokenize(text.lower())
```

# 3. Remove Punctuation

```python
punctuation
```

```python
# word tokenisasi
tokens = word_tokenize(text.lower())


# remove punctuation
new_tokens = [item for item in tokens if item not in punctuation]
new_tokens
```

# 4. Remove Stopwords
Stopwords adalah kata-kata yang diabaikan dalam NLP karena tidak memiliki makna yang signifikan.

```python
list_sw_nltk     = stopwords.words("indonesian")             # ada 758
list_sw_sastrawi = StopWordRemoverFactory().get_stop_words() # ada 126
```

Stopwords dari NLTK lebih kejam (terlalu banyak kata dari Bahasa yang dianggap tidak punya makna) dibandingkan dengan Sastrawi.

```python
# remove stopwords
new_tokens_v2 = [item for item in new_tokens if item not in list_sw_sastrawi]
new_tokens_v2
```
