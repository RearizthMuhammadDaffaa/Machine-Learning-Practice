# 1. Import Packages

```python
import numpy as np
import pandas as pd

from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from string import punctuation

!pip install Sastrawi
from Sastrawi.StopWordRemover.StopWordRemoverFactory import StopWordRemoverFactory
stopwords_sastrawi = StopWordRemoverFactory()
stopwords_sastrawi = stopwords_sastrawi.get_stop_words()

list_sw_punct = stopwords_sastrawi + list(punctuation)
```

# 2. Extract Data

```python
df = pd.read_csv('/content/drive/MyDrive/Course Files/Course-Net/02. Kampus Merdeka/10. Natural Language Processing/kompas.csv')
df.head()
```

# 3. Word Embedding

```python
from sklearn.feature_extraction.text import TfidfVectorizer

import nltk
nltk.download('punkt')
```

```python
tfidf_model  = TfidfVectorizer(tokenizer=word_tokenize,stop_words=list_sw_punct)
tfidf_matrix = tfidf_model.fit_transform(df['teks'])
```

```python
tfidf_matrix.shape
```

# 4. Extract Keywords

```python
# vocab
vocab = tfidf_model.get_feature_names_out()
vocab.shape
```

```python
## vocab tertentu
vocab[17619]
```

```python
## extract top n keyword
index = 0
top_n = 10

sorted_index = tfidf_matrix.toarray()[index,:].argsort()[::-1]
[vocab[index] for index in sorted_index[:top_n]]
```

# 5. Document Similarity

```python
## Menghitung similarity antara 2 vektor
kalimat_1 = np.array([0,1,2,9,4,5,3]).reshape(1,-1 )
kalimat_2 = np.array([0,1,2,7,4,5,2]).reshape(1,-1 )

from sklearn.metrics.pairwise import cosine_similarity

## kedua vektor itu sama jika nilai cosine_similaritynya menuju 1
cosine_similarity(kalimat_1,kalimat_2)
```

```python
index = 2000
top_n = 3
print(df.loc[index,:])


similarity_value = cosine_similarity(tfidf_matrix[index],tfidf_matrix)
sorted_index     = similarity_value.argsort().reshape(-1,)[::-1][1:]

print('')
print('Top N artikel yang paling Similar')
print(40*'=')
for i in range(top_n):
    print(i+1,'. Similarity Value :',similarity_value[:,sorted_index[i]])
    print(df.loc[sorted_index[i],:])
    print('')
```
