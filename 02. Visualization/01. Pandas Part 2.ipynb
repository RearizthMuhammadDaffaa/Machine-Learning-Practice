# 4. Handling Missing Value
Ada 2 pendekatan dalam _handling missing value_, diantaranya :
1. drop (baris/kolom yang punya missing value)
2. impute

Dalam hal ini, kita akan impute missing value dengan nilai 0.
<br>Asumsinya adalah produk tersebut tidak mengandung kandungan variabel tersebut, yang artinya nilainya = 0.

```python
# Impute
df['carbo']  = df['carbo'].fillna(0)
df['sugars'] = df['sugars'].fillna(0)
df['potass'] = df['potass'].fillna(0)


# # Drop : baris yang punya missing value
# df = df.dropna()
```

# 9. Encoding & Decoding
- encoding : kategorikal -> numerikal
- decoding : numerikal -> kategorikal

```python
#  Encoding Type 1
## OneHotEncoder
pd.get_dummies(df,dtype=int).head()
```

```python
df['type'].unique()
```

```python
#  Encoding Type 2
## LabelEncoder
mapper = {'C' : 1,
          'H' : 2}

df['type'] = df['type'].map(mapper)
```

```python
df = pd.get_dummies(df,dtype=int)
```

```python
## Decoding
## Cara :
## 1. decoding berdasarkan mapper-nya
## 2. binning ke kategorinya

## binning rating : klasifikasi ratingnya menjadi [buruk,rendah,tinggi,etc.]

df['class_rating'] = pd.cut(df['rating'],bins=[18,50,60,70,80,100],
                            labels=['buruk','rendah','cukup','baik','sangat baik'])
```
